```python
it_it_matrix = pd.DataFrame(index=df_adj.columns,columns=df_adj.columns)
it_it_matrix.head(5)
```

```python
for i in range(0,len(it_it_matrix.columns)) :
# Loop through the columns for each column
    for j in range(0,len(it_it_matrix.columns)) :
    # Fill in placeholder with cosine similarities
        it_it_matrix.iloc[i,j] = 1-cosine(df_adj.iloc[:,i],df_adj.iloc[:,j])
```

```python
it_it_matrix.head(60)
```

```python
df_rank = pd.DataFrame(index=it_it_matrix.columns,columns=range(1,11))
for i in range(0,len(it_it_matrix)):
    df_rank.iloc[i, :10]=it_it_matrix.iloc[0:,i].sort_values(ascending=False)[:10].index
    
df_rank.head(10)
```
